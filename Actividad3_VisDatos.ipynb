```python
import pandas as pd
import numpy as np
data='https://raw.githubusercontent.com/tec03/Datasets/main/datasets/antropometria-dataset.csv'
```

```python
df=pd.read_csv(data)
df.head()
```

```python
ndf=df.copy()
```

```python
pd.pivot_table(index=['desc_ent','sexo'],aggfunc=['median','mean'],data=ndf,values= 'peso',margins=True)
```
