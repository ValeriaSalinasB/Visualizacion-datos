```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path='/content/drive/MyDrive/Act7_VisDatos_A01284888.xlsx'
```

```python
df=pd.read_excel(path,
                 sheet_name='Sheet1')
df
```

```python
sns.histplot(data=df['cintura'], bins=15)
```

```python
sns.lmplot(x='diastol',
           y='cintura',
           data=df,
           hue='sexo',
           fit_reg=False)
```

```python
ms=df.loc[df['sexo']==1,'cintura'].to_frame()
```

```python
fs=df.loc[df['sexo']==2,'cintura'].to_frame()
```

```python
fig=plt.figure()
ax1=fig.add_subplot(1,3,1)
sns.boxplot(data=ms)
ax2=fig.add_subplot(1,3,3) #1 row, 3 columns, position 3
sns.boxplot(data=fs)

ax1.title.set_text('Male score')
ax2.title.set_text('Female score')
```

```python
mujer_sistol=pd.pivot_table(index=['desc_ent','sexo'],data=df,values= 'sistol' ,margins=True)
mujer_sistol
```

```python
mujer_diastol=pd.pivot_table(index=['desc_ent','sexo'],data=df,values= 'diastol' ,margins=True)
mujer_diastol
```

```python
sns.lmplot(x='sistol',
           y='diastol',
           data=df,
           hue='sexo',
           fit_reg=False)
```
